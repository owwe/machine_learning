```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
from ml import KMeans
# Generate sample data
n_samples = 100
n_components = 4

X, y_true = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=0.60, random_state=0
)
X = X[:, ::-1]
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis')
```

```python
k_means = KMeans(n_components,max_iter= 200)
k_means.fit(X)
```

```python
x_test, y_test = make_blobs(
    n_samples= 20, centers=n_components, cluster_std=0.60, random_state=0
)
x_test = x_test[:, ::-1]
ax = plt.subplot(211)
ax.set_title('ground truth')
ax = plt.scatter(x_test[:,0],x_test[:,1],c = y_test, cmap='viridis')

ax2 = plt.subplot(212)
ax2.set_title('k-means results')
preds = k_means.predict(x_test)
ax2 = plt.scatter(x_test[:,0],x_test[:,1],c = preds, cmap = 'Pastel2')
ax2 = plt.scatter(k_means.centroids[:,0],k_means.centroids[:,1],marker = 'x',c = 'k')
```
